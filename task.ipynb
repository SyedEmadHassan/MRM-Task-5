```python
df = pd.read_csv("CarPrice_Assignment.csv")
df['intercept'] = 1

```

```python

df_encoded = pd.get_dummies(df, columns=['fueltype'], prefix='fuel')
df_encoded
```

```python
df_new= pd.get_dummies(df_encoded, columns=['carbody'], prefix='body')
print(df_new)
```

```python
df_new = pd.get_dummies(df_new,columns=["doornumber"],prefix="doors")
print(df_new)
```

```python
df_new=pd.get_dummies(df_new,columns=["drivewheel"])
print(df_new)
```

```python
df_new=pd.get_dummies(df_new,columns=["enginelocation"],prefix="engineloc")
df_new=pd.get_dummies(df_new,columns=["enginetype"])
df_new = pd.get_dummies(df_new,columns=["cylindernumber"],prefix="cylnum")
df_new = pd.get_dummies(df_new,columns=["fuelsystem"],prefix="fuelsys")
df_new = pd.get_dummies(df_new,columns=['aspiration'],prefix='asp')

print(df_new)
```

```python
test = df_new*1
test.drop(['car_ID','CarName'],axis=1,inplace=True)
test
```

```python

def standardize(arr):
    means = np.mean(arr,axis=0)
    std_devs= np.std(arr,axis=0)
    std_devs[std_devs==0]=1
    std_arr = (arr-means)/std_devs
    return std_arr

x = test.drop(['price'],axis=1)
y= test['price']
no_hot_x=x.iloc[:,:12]
no_hot_x=standardize(no_hot_x) #standardising without binary columns
x = pd.concat([no_hot_x,x.iloc[:,12:]],axis=1)
x = x.to_numpy()
```

```python
def cost(data,weights,targets):
    total_cost = 0
    for i in range(len(data)):
        total_cost+=(1/len(data))*((data[i]*weights).sum()-targets[i])**2
    return total_cost
```

```python
def g_descent(data,weights,targets,learn_rate,iter):

    for i in range(iter):
        gradients= np.zeros(53)
        for j in range(len(data)):
            for k in range(53):
                   gradients[k] += (1/len(data))*((data[j]*weights).sum()-targets[j])*data[j,k]
        weights=weights - learn_rate*gradients
        print(cost(data,weights,targets))
    return weights
```

```python
w = np.zeros(53)
w_final = g_descent(x,w,y,0.001,1000)
```
